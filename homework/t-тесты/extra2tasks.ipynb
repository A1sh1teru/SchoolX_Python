```python
import scipy.stats as stats
```

###  Изменилось ли время использования личного кабинета на сайте после редизайна?

$H_0$: `Нет` статистически значимых различий во времени использования личного кабинета до и после редизайна сайта. <br>
$H_1$: `Есть` статистически значимые различия во времени использования личного кабинета до и после редизайна сайта.

```python
time_before = [1732, 1301, 1540, 2247, 1632, 1550, 754, 1946, 1889,
          2748, 1349, 1648, 1665, 2416, 1470, 1681, 1868, 1629,
          1271, 1633, 2131, 942, 1599, 1127, 2200, 661, 1207,
          1737, 2410, 1486]

time_after = [955, 2577, 360, 139, 1618, 990, 644, 1796, 1487, 949, 472,
         1906, 1758, 1258, 2554, 612, 309, 1864, 1294, 1487, 1164, 1559,
         491, 2286, 1270, 2069, 1553, 1629, 1704, 1623]

t_statistic, p_value = stats.ttest_rel(time_after, time_before)

alpha = 0.05

if p_value < alpha:
    result = "Отвергаем нулевую гипотезу. Имеются статистически значимые различия во времени использования личного кабинета после редизайна."
else:
    result = "Не отвергаем нулевую гипотезу. Нет статистически значимых различий во времени использования личного кабинета после редизайна."

print(f'p-value: {p_value}')

result
```

###  Изменилось ли количество используемых патронов игроками после нововведений в онлайн-игре?

$H_0$: `Нет` статистически значимых различий в количестве используемых патронов до и после внесенных изменений в онлайн-игре. <br>
$H_1$: `Есть` статистически значимые различия в количестве используемых патронов до и после внесенных изменений в онлайн-игре.

```python
bullets_before = [821, 1164, 598, 854, 455, 1220, 161, 1400, 479, 215,
          564, 159, 920, 173, 276, 444, 273, 711, 291, 880,
          892, 712, 16, 476, 498, 9, 1251, 938, 389, 513]

bullets_after = [904, 220, 676, 459, 299, 659, 1698, 1120, 514, 1086, 1499,
         1262, 829, 476, 1149, 996, 1247, 1117, 1324, 532, 1458, 898,
         1837, 455, 1667, 898, 474, 558, 639, 1012]

t_statistic, p_value = stats.ttest_ind(bullets_after, bullets_before)

alpha = 0.05

if p_value < alpha:
    result = "Отвергаем нулевую гипотезу. Имеются статистически значимые различия в количестве используемых патронов после внесенных изменений."
else:
   result = "Не отвергаем нулевую гипотезу. Нет статистически значимых различий в количестве используемых патронов после внесенных изменений."

print(f'p-value: {p_value}')

result
```
