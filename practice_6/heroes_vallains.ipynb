```python
class Character:
    strength: int | float
    agility: int | float
    intelligence: int | float
    speed: int | float
    damage: int | float

    def __init__(
        self, 
        strength,
        agility,
        intelligence,
    ):
        self.strength = strength
        self.agility = agility
        self.intelligence = intelligence
        self.speed = agility * 2.5
        self.damage = strength * 1.5

    def trade(self):
        print('Склонен торговать')
    def wait(self):
        print('Ждет')
    def think(self):
        print('Мудро рассуждает')
    def move(self):
        print('Передвигается')

class Hero(Character):
    kindness: int | float
    politeness: int | float
    
    def __init__(self, strength, agility, intelligence, kindness, politeness):
        super().__init__(strength, agility, intelligence)
        self.kindness = kindness
        self.politeness = politeness

    def protecting(self):
        print('Защищает')

    def joking(self):
        print('Шутит')

class Villain(Character):
    evilness: int | float
    rudeness: int | float

    def __init__(self, strength, agility, intelligence, evilness, rudeness):
        super().__init__(strength, agility, intelligence)
        self.evilness = evilness
        self.rudeness = rudeness

    def murder(self):
        print('Совершает убийства')

    def lie(self):
        print('Врёт')

class Policeman(Hero):
    def __init__(self, strength, agility, intelligence, kindness, politeness, courage):
        super().__init__(strength, agility, intelligence, kindness, politeness)
        self.courage = courage

    def arrest(self):
        print('Арестовывает')

    def wears_unifrom(self):
        print('Носит форму')

class Killer(Villain):
    def __init__(self, strength, agility, intelligence, evilness, rudeness, cruelty):
        super().__init__(strength, agility, intelligence, evilness, rudeness)
        self.cruelty = cruelty

    def uses_weapon(self):
        print('Использует оружие')

    def wears_gloves(self):
        print('Носит перчатки')
```

```python
policeman_1 = Policeman(30, 15, 20, 25, 25, 40)
policeman_2 = Policeman(25, 20, 15, 45, 30, 35)
killer_1 = Killer(35, 30, 15, 10, 40, 45)
killer_2 = Killer(45, 20, 25, 30, 20, 35)

policeman_1.move()
policeman_2.arrest()
killer_1.uses_weapon()
killer_2.wears_gloves()
```
